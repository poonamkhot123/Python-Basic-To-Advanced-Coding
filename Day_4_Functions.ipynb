```python
def is_prime(num):
  count = 0
  for i in range(1,num+1):
    if num % i == 0:
      count += 1
    if count == 2:
      return True
    else:
      return False
```

```python
is_prime(45)
```

```python
is_prime(17)
```

```python
is_prime(23)
```

```python
# Write a function called min_of_sqaure that takes three numbers and returns the smallest
# without using any built in function.
def min_of_square(num1,num2,num3):
  if num1 < num2 and num1 < num3:
    return num1

min_of_square(10,20,30)
```

```python
def min(a,b,c):
  if a > b and a > c:
    return c
  elif b > a and b > c:
    return a
  elif c > a and c > b:
    return c

result = min(5,7,9)
print(result)
```

```python
def absolute_value(num):
  if num < 0:
    return -num
  else:
    return num

result = absolute_value(-10)
print(result)
```

```python
# keyword Arguments.
def calculate_marks(maths,eng=0,hindi=0):
  print(f"maths marks {maths}")
  print(f"english marks {eng}")
  print(f"hindi marks {hindi}")

calculate_marks(maths=100)
```

```python
# Local Variable = Variable inside the function.
def addition(n1,n2,n3):
  total = n1+n2+n3
  print(f"The total is {total}")

addition(10,20,56)
```

```python
# Global Variable.
name = "Anirudhha"

def greet():
  name = "Muskan"
  print(f"Hey {name}! Good Morning")

greet()
print(name)
```

```python
count = 0
def increase():

  count = 1
  print(f"Inside function count = {count}")

increase()
print(f"Outside function count= {count}")

```

```python
count = 0
def increase():
  global count
  count = 1
  print(f"Inside function count = {count}")

increase()
print(f"Outside function count= {count}")
```

```python
# Lambda Function = Anonymous Function written in a single line.
square = lambda num : num * num
print(square(5))
```

```python
sum = lambda a,b,c : a+b+c
print(sum(10,56,45))
```

```python
a = lambda age : True if age >= 18 else False
print(a(45))
```

```python
# write a lambda function that takes a number and returns its cube. Store it in a
# variable and call it.
x = lambda cube : cube*cube*cube
print(x(5))
```

```python
# Write a lambda function that takes a number and returns " Positive"
# or " Negative"
sum = lambda num : "Positive" if num > 0 else "Negative"
print(sum(-5))
```

```python
# write a function fizzbuzz(n) that takes a single number and prints 'Fizz"
# if its divisible by 3,"Buzz"if its divisible by 5 ,
# "FizzBuzz" if its divisible by both, otherwise print the number itself.
def fizzbuzz(n):
  if n % 3 == 0:
    print ("Fizz")
  elif n % 5 == 0:
    print("Buzz")
  elif n % 5 == 0 and n % 3 == 0:
    print("FizzBuzz")
  else:
    return n
```

```python
fizzbuzz(7)
```

```python
fizzbuzz(15)
```

```python
fizzbuzz(20)
```

```python
fizzbuzz(30)
```
