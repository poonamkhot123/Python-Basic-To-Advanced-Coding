# type conversion

```python
# 1 .implicit - python does it automatically
x = 5
y = 2.0
z = x+y
print(z)
print(type(z))
```

```python
# 2.Explicit - you do it manually
num_str = "42"
num_int = int(num_str)
num_flt = float(num_int)
print(num_flt)
```
