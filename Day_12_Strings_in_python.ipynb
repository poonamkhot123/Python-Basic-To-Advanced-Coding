```python
# Clean Phone Number.
```

```python
def clean_number(phone_number:str):
  phone_number = phone_number.replace("-","")
  phone_number = phone_number.replace(" ","")
  phone_number = phone_number.replace("(","")
  phone_number = phone_number.replace(")","")
  return phone_number

print(clean_number("123-456-7890"))
```

```python
# vowel start words.
sentence = "poonam is an excellent coder and a great engineer."
def count_vowel_words(sentence:str):
  vowels = "aeiouAEIOU"
  count = 0
  words = sentence.split()
  for word in words:
    if word[0] in vowels:
      count += 1
  return count

print(count_vowel_words(sentence))
```

```python
# longest word.
sentence = "poonam is a pro coder and a programmer"
def print_longest_word(sentence:str):
  words = sentence.split()
  longest_count = 0
  longest_word = ""
  for word in words:
    if len(word) > longest_count:
      longest_count = len(word)
      longest_word = word
  print(longest_word)

  print(words)


print_longest_word(sentence)
```

```python
# username validation.
def validate_username(username):
  username = username.strip()
  if username[0].isdigit():
    return "Invalid"
  return "valid"

username = "   Aniruddha"
print(validate_username(username))
```

```python
# unique words count.
def count_unique_words(paragraph):
  words = paragraph.split()
  print(set(words))
```

```python
paragraph = "Python is great and python is fun and learning is the best."
print(count_unique_words(paragraph))
```

```python
def is_palindrome(text):
  text = text.lower().replace(" ","")
  return text == text[::-1]
```

```python
print(is_palindrome("poonamkhot"))
```

```python
print(is_palindrome("Never odd or even."))
```

```python
# Reverse word Order.
def reverse_words(sentence:str):
  words = sentence.split()
  words = words[::-1]
  return " ".join(words)


sentence = "Aniruddh is learning  Python programming every single day."
print(reverse_words(sentence))
```
