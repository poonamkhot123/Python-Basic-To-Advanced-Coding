```python
# Abstraction : Hiding complexity behind clean interfaces.
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
  def perimeter(self):
    pass
class Rectangle(Shape):
  def __init__(self,l,b):
    self.l = l
    self.b = b

  def area(self):
    return self.l * self.b

  def perimeter(self):
    return 2 * (self.l + self.b)

rect = Rectangle(10,20)
rect.area()
```

```python
rect.perimeter()
```
