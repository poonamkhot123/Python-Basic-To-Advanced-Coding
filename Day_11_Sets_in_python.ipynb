# Set operations : union and intersection.

```python
a = {1,2,3,4}
b = {3,4,5,6}
print(a|b)
```

```python
print(a.union(b))
```

```python
a = {1,2,3,4}
b = {3,4,5,6}
print(a & b)
```

```python
print(a.intersection(b))
```
