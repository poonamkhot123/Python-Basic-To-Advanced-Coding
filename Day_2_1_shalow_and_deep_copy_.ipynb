```python
import copy
original = [1,2,3]
shallow = copy.copy(original)

print(id(original))
print(id(shallow))

shallow.append(100)
```

```python
print(original)
```

```python
print(shallow)
```

```python
import copy

original = [54,12,76,[67,19,20,78],98,100,3]

#shallow = original.copy()
shallow = copy.deepcopy(original)

print(id(original))
print(id(shallow))

shallow[3][1] = 999
shallow[6] = 1000

print(original)
print(shallow)
```

# Pass by value


```python
def add_number(x):
  x = x + 1
  print(f"Inside function = {x}")

num = 10
add_number(num)
print(f"outside function = {num}")
```

# pass by reference


```python
def add_item(x):
  x.append(100)
  print(f"Inside function = {x}")

nums = [1,2,3]
add_item(nums)

print(f"outside function = {nums}")
```

```python
def rebind(x):
  x = [100,200,300]
  print(f"inside function = {x}")

nums = [1,2,3]
rebind(nums)
print(f"outside function = {nums}")
```

```python
def mutate(x):
  x[0] = 100
  print(f"Inside function = {x}")

nums = [1,2,3]
mutate(nums)
print(f"Outside function = {nums}")
```
