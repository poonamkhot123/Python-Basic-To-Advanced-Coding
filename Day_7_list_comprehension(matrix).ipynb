```python
# using list comprehension, create a list of squares of all odd numbers from 1 to 20.
list = [i*i for i in range(1,21) if i%2 !=0]
```

```python
list
```

```python
# Given a list of marks,use list comprehension
#  to create a new list that contains only the marks that are above 75.
marks = [23,34,12,94,37,89,99,32,11]
new_list = [i for i in marks if i >= 75]
```

```python
matrix = [[5,7,2,3,1],
          [6,3,8,1,3],
          [6,4,9,0,2],
          [3,6,3,8,1]]
rows = len(matrix)
columns = len(matrix[0])
for i in range(0,rows):
  for j in range(0,columns):
    print(matrix[i][j],end = " ")
  print()
```

```python
matrix = [
    [1,2,3,1],
    [4,5,6,3],
    [7,8,9,4],
    [5,6,7,8]
]

r = len(matrix)
c = len(matrix[0])

for i in range(0,r):
  for j in range(0,c):
    if i>= j:
      print(matrix[i][j],end = " ")
    else:
      print("*",end = " ")
  print()
```

```python
matrix = [
    [3,4,5,6],
    [6,7,8,9],
    [7,8,9,6],
    [5,6,7,8]
]
r = len(matrix)
c = len(matrix[0])
for i in range(0,r):
  for j in range(0,c):
    if i +j == r-1:
      print(matrix[i][j],end = " ")
    else:
      print("*",end = " ")
  print()
```
