```python
marks = {
    "akshay" : [43,12,98,10,22],
    "priya" :  [76,55,34,89,61],
    "rahul" :  [20,45,67,30],
    "sneha" :  [91,73,50,14,62],
    "karan" :  [38,82,47,95,29]
}
```

```python
# sort the list according to the last marks.
dict(sorted(marks.items(),key= lambda x:x[1][-1]))
```

```python
# sorted list with the sum of all marks.
dict(sorted(marks.items(),key=lambda x:sum(x[1])))
```

```python
# sort the list according to descending order.
ans = dict(sorted(marks.items(),key = lambda x : sum(x[1]),reverse = True))
```

```python
print(ans)
```

# sorted nested dictionary.

```python
students = {
    "anirudhha": {"math":87,"science":92,"english":74},
    "priya" :    {"math":65,"science":78,"english":90},
    "rahul" : {"math":55,"science":60,"english":82},
    "sneha" : {"math":95,"science":88,"english":82},
    "karan" : {"math":70,"science":45,"english":63}
}
```

```python
ans = dict(sorted(students.items(),key=lambda x:x[1]["math"]))
```

```python
print(ans)
```

```python
# sort on the basis of total amrks scored.
ans = dict(
    sorted(
        students.items(),key = lambda x:x[1]['math'] + x[1]['science']+x[1]['english']
    )
)
```

```python
print(ans)
```

# dictionary Comprehension

```python
squares = {}
for i in range(1,11):
  squares[i] = i*i
print(squares)
```

```python
squares = {i:i for i in range(1,6)}
print(squares)
```

```python
# with a condition
marks = {"math":85,"science":92,"english":60,"hindi":45}
```

```python
top = {sub: m for sub, m in marks.items() if m > 80}
print(top)
```

```python
students = {
    "akshay" : [43,12,98,10,22],
    "priya" :  [76,55,34,89,61],
    "rahul" :  [20,45,67,30,88],
    "sneha" :  [91,73,50,14,62],
    "karan" :  [38,82,47,95,29]
}
```

```python
for name,marks in students.items():
  total = sum(marks)
  avg = total / len(marks)
  print(f"{name} has scored total of {total} marks with avg of { avg:.2f} marks.")
```
