```python
class Student:
  # Attributes.
  roll_no = 0
  name = " "
  gender = " "
  age = 0

  def set_details(self):
    self.roll_no = int(input("Enter roll no."))
    self.name = input("Enter name =")
    self.gender = input("Enter gender")
    self.age = int(input("Enter a age"))

  # method
  def display_details(self):
    print(f"roll no = {self.roll_no}")
    print(f"name = {self.name}")
    print(f"Gender = {self.gender}")
    print(f"age = {self.age}")


# object/instance
student1 = Student()
student1.set_details()
student1.display_details()

student2 = Student()
student2.set_details()
student2.display_details()
```

```python
class Student:
  # Attributes.
  roll_no = 0
  name = " "
  gender = " "
  age = 0

  # methods
  def set_details(self,r:int,n:str,g:str,a:int):
    self.roll_no = r
    self.name = n
    self.gender = g
    self.age = a

student1 = Student()
student1.set_details(1,"Aniruddha","male",88)
print(student1.roll_no)
print(student1.name)
print(student1.gender)
print(student1.age)
```

```python
class student:
  def __init__(self,name:str,age:int,marks:list[int]):
    self.name = name
    self.age = age
    self.marks = marks

  def total(self):
    return sum(self.marks)

  def average(self):
    return sum(self.marks) / len(self.marks)

  def grade(self):
    avg = sum(self.marks)/len(self.marks)
    if avg >=90:
      print("A")
    elif avg >= 50 and avg  < 89:
      print("B")
    else:
      print("c")

student1 = student("poonam",67,[56,4,5,6,78,34])
print(student1)
avg = student1.average()
print(avg)
student1.grade()
```

```python
total = student1.total()
print(total)
```
