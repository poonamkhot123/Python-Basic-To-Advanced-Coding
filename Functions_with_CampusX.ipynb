```python
def is_even(num):
  """THis function returns if a given number is odd or even
  input - any valid integer
  output - odd/even
  created on - 4/01/2026
  """
  if num % 2==0:
    return "even"
  else:
    return"odd"
```

# *args and **kwargs

*args : allows us to pass a variable number of non-keyword arguments to a function.

```python
def multiply(*args):
  product = 1

  for i in args:
    product = product * i
  return product
```

```python
multiply(12,3,4,9,5)
```

```python
# how to see documentation of any function.
print(functionname.__doc__)
```

```python
# **kwargs
# **kwargs allows us to pass any number of keyword arguments.
# keyword arguments mean that they contain a key-value pair,like a Python dictionary.

def display(**kwargs):
  for (key,value) in kwargs.items():
    print(key,"->",value)
```

```python
display(india='delhi',shrilanka='colombo',nepal='kathmandu')
```

# Lambda Function.
* A lambda function is a small anonumous function.
* A lambda function can take any number of arguments,but can only have one expression.

```python
a = lambda x:x**2
a(2)
```

```python
b = lambda x,y:x+y
b(2,3)
```

# Difference between lambda vs normal function
* No name
* lambda has no return value(infact ,returns a function)
* lambda is written in 1 line.
* not reusable.

# Then why used lambda fuction?
* They are used with Higher order Functions.

```python
# check if a string has "a"
a = lambda s: "a" is s
a("hello")
```

```python
# odd or even
a = lambda x:'even' if x%2 ==0 else 'odd'
a(6)
```

# Higher order function.
* A Higher-Order Function (HOF) in Python is a function that does at least one of the following:

* Takes one or more functions as arguments
* Returns a function as its result

# map

```python
# square the items of a list.
list(map(lambda x:x**2,[1,2,3,4,5]))
```

```python
# odd/even labelling of list items.
L = [1,2,3,4,5]
list(map(lambda x :'even' if x%2==0 else 'odd',L))
```

```python
# fetch names from a list of dict.
users = [
  {
   'name':'rahul',
   'age':45,
   'gender':'male'
   },
  {
   'name':'nitish',
   'age':33,
   'gender':'male'
   },{
   'name':'poonam',
   'age':23,
   'gender':'female'
   }
    ]
```

```python
list(map(lambda users:users['gender'],users))
```

# Filter

```python
# numbers greater than 5
L = [3,4,5,6,7]
list(filter(lambda x: x>5, L))
```

```python
# fetch fruits with 'a'.
fruits = ['apple','guava','cherry']
list(filter(lambda x:x.startswith('a'),fruits))
```

# Reduce.

```python
# sum of all item
import functools
functools.reduce(lambda x,y:x+y,[1,2,3,4,5])
```

```python
# find minimum number
functools.reduce(lambda x,y:x if x<y else y,[23,11,45,10,9])
```

```python
# find maximum number
functools.reduce(lambda x,y:x if x>y else y,[23,11,45,10,9])
```
