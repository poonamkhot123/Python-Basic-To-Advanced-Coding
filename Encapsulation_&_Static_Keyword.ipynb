# Write OOP classes to handle the following scenarios.
* 1) A user can create and view 2D coordinates.
* 2) A user can find out the distance between 2 coordiantes.
* 3) A user can find the distance of a coordinate from origin.
* 4) A user can check if a point lies on a given line.
* 5) A user can find the distance between a given 2D point and a given line.

```python
class Point:
  def __init__(self,x,y):
    self.x_cod = x
    self.y_cod = y

  def __str__(self):
    return "<{},{}>".format(self.x_cod,self.y_cod)

  def euclidean_distance(self,other):
    return ((self.x_cod-other.x_cod)**2 + (self.y_cod-other.y_cod)**2)**0.5

  def distance_from_origin(self):
    # return (self.x_cod**2 +self.y_cod**2)**0.5
    return self.euclidean_distance(Point(0,0))
```

```python
p1 = Point(1,1)
print(p1)
```

```python
p2 = Point(2,4)
print(p2)
```

```python
p1.euclidean_distance(p2)
```

```python
p2.distance_from_origin()
```

```python
class Line:
  def __init__(self,A,B,C):
    self.A = A
    self.B = B
    self.C = C

  def __str__(self):
    return '{}x+{}y+{}=0'.format(self.A,self.B,self.C)

  def point_on_line(line,point):
    if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
      return "lies on the line"
    else:
      return "does not lie on the line."

  def shortest_distance(line,point):
    return abs(line.A*point.x_cod + line.B*point.y_cod + line.C)/(line.A**2 +line.B**2)**0.5
```

```python
l1 = Line(3,4,5)
print(l1)
```

```python
l1 =Line(1,1,-2)
p1 = Point(1,1)
print(l1)
print(p1)

l1.point_on_line(p1)
```

```python
l1 =Line(1,1,-2)
p1 = Point(1,10)
print(l1)
print(p1)

l1.shortest_distance(p1)
```

# How object access attributes.

```python
class Person:
  def __init__(self,name_input,country_input):
    self.name = name_input
    self.country = country_input
  def greet(self):
    if self.country == "india":
      print("Namste",self.name)
    else:
      print("Hello",self.name)
```

```python
# how to access attributes.
p = Person('nitish','india')
```

```python
p.name
```

```python
# how to access methods
p.greet()
```

# Reference Variables
* Reference variables hold the objects
* We can create objects without reference variable as well.
* An object can have multiple reference variables.
* Assiging a new reference variable to an existing object not create a new proejct.

# pass by reference

```python
class Person:
  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class -> function.
def greet(person):
  print("hI my name is",person.name,'and i am a',person.gender)

p = Person('nitish','male')
greet(p)
```

# what is instance variable.

```python
# dict of object
class Person:
  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

p1 = Person('nitish','male')
p2 = Person('ankita','female')
p3 = Person('sumit','male')

L = [p1,p2,p3]

for i in L:
  print(L)
  print(i.name)
  print(i.gender)
```
