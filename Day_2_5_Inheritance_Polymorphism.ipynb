# Super() method = if we want to write both.

* if the name is same of both the functions.

* if we want information of parent class in the child class then we use super() method.

```python
class Animal:
  def speak(self):
    print("Animal is speaking")

class Dog(Animal):
  def speak(self):
    super().speak()
    print("dog is barking")

d = Dog()
d.speak()
```

```python
class Vehicle:
  def __init__(self,brand):
    print("This is vehicle constructor.")
    self.brand = brand

class Car(Vehicle):
  def __init__(self,fuel,brand):
    print("This is car constructor")
    super().__init__(brand)
    self.fuel = fuel


car = Car("petrol","maruti")

```

# Diamond Problem.
* Multiple inheritance with a shared ancestor creates ambiguity.

* It will solve by using MRO Method resolution order.



```python
class A:
  def hello(self):
    print("Hello from A")

class B(A):
  def hello(self):
    print("Hello from B")

class C(A):
  def hello(self):
    print("hello from C")

class D(B,C):
  pass

d = D()
d.hello()
```

# Polymorphism : many forms.
* The same operations behaves differently for different types.

```python
class Dog:
  def speak(self):
    print("woolf")

class Cat:
  def speak(self):
    print("Meow!")

class Cow:
  def speak(self):
    print("Moo!")

animals =[Dog(),Cat(),Cow()]
for animal in animals:
  animal.speak()
```
