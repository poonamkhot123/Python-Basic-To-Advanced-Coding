# List FUnctions


```python
# len/min/max/sorted
L = [2,1,5,7,0]
print(len(L))
print(min(L))
print(max(L))
print(sorted(L))
print(sorted(L, reverse=True))
```

```python
# count
L = [1,2,1,4,3,1,5]
L.count(1)
```

```python
# index
L = [1,2,1,4,3,1,5]
L.index(1)
```

```python
# reverse
L = [1,2,1,4,3,1,5]
# permanently reverse list
L.reverse()
print(L)
```

```python
# sort vs sorted
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
```

```python
# copy  -> shallow
L = [2,1,5,7,0]
print(L)
print(id(L))

L1 = L.copy()
print(L1)
print(id(L1))
```

# List Comprehension
* List COmprehension provides a concise way of creating lists.
* newlist = [expression for item in iterable if condition == True]
* Advantages
* More time - efficient and space - efficient than loops.
* Require fewer lines of code.
* Transforms iterable statement into a formula.


```python
# what is deep vs shallow
```

```python
# add 1 to 10 numbers to a list.
L  = []
for i in range(1,11):
  L.append(i)
print(L)
```

```python
L =  [ i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3

[s*i for i in v]

print(v*s)
```

```python
# add squares
L = [1,2,3,4,5]

[i**2 for i in L]
```

```python
# print all numbers divisible by 5 in the range of 1 to 50.
[ i for i in range(1,51) if i%5 == 0]
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

[language for language in languages if language.startswith("p")]
```

```python
# Nested if with List Comprehension
basket = ["apple",'Guava',"cherry","banana"]
my_fruits = ["apple","kiwi","grapes","banana"]

# add new list from my_fruits and items if the fruit exists in basket ans also starts with "a".

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith("a")]
```

```python
# print a (3,3) matrix using list comprehension -> Nested List comprehension.
[[i*j for i in range(1,4)]for j in range(1,5)]
```

```python
# Cartesian products - > List comprehension on 2 lists together.
L1 = [1,2,3,4]
L2 = [4,5,6,7]
[i*j for i in L1 for j in L2]
```

# 2 ways to traverse a list.
* itemwise
* indexwise

```python
# itemwise
L = [1,2,3,4]

for i in L:
  print(i)
```

```python
# indexwise : it return a index of each element.
L = [1,2,3,4]
for i in range(0,len(L)):
  print(i)
```

```python
L = [1,2,3,4]
for i in range(0,len(L)):
  print(L[i])
```

# Zip
* The zip() function returns a zip object,which is an iterator where the first item in each passed is paired together,and then the second item in each passed iterator are paired toghether.
* if the passed iterators have different lengths, the iterator with the least items decides the length of the new ieterator.

```python
# write a program to add items of 2 lists indexwise.
L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))
```
