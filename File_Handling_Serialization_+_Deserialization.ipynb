# using Context Manager(with)
* it's a good idea to close a file after usage as it will free up the resources.
* If we don't close it,garbage collector close it.
* with keyword closes the file as soon as the usage is over.

```python
# with
with open("/content/Temp/sample.txt",'w') as f:
  f.write("salman bhai")
```

```python
# try f.read() now
with open("/content/Temp/sample.txt") as f:
  print(f.read())
```

```python
# moving within a file -> 10 char then 10 char.
with open("/content/Temp/sample.txt","r") as f:
  print(f.read(10))
  print(f.read(10))
```

```python
# benefit ? -> to load a big file in memory.
big_L = ["hello world" for i in range(1000)]

with open("big.txt","w") as f:
  f.writelines(big_L)
```

```python
with open("big.txt","r") as f:
  chunk_size = 10

  while len(f.read(chunk_size)) > 0:
    print(f.read(chunk_size),end = "***")
    f.read(chunk_size)
```

```python
# seek and tell function.
with open("/content/Temp/sample.txt") as f:
  print(f.read(10))
  print(f.tell())
  f.seek(15)
  print(f.read(10))
  print(f.tell())
```

```python
# seek during write.
with open("/content/Temp/sample.txt","w") as f:
  f.write("hello")
  f.seek(0)
  f.write("x")
```

# Serialization and Deserialization
* Serialization - process of converting python data types to JSON ( Java Skript On Notation)  formate.
* Deserialization - process of converting JSON  to python data types.

```python
# serialization using JSON module.
# list
import json
L = [1,2,3,4]
with open("demo.json",'w') as f:
  json.dump(L,f)
```

```python
# dict
d ={
    "name":"nitish",
    "age":33,
    "gender":"male"
}

with open("demo.json","w") as f:
  json.dump(d,f,indent =4)
```

```python
# deserialization.
import json

with open("demo.json","r") as f:
  d = json.load(f)
  print(d)
  print(type(d))
```

```python
# srerialize and deserialize tuple.
import json
t = (1,2,3,4)

with open("demo.json","w") as f:
  json.dump(t,f)
```

```python
# serialize and deserialize a nested dict.
d = {
    "student":'nitish',
    "marks":[23,14,34,45,56]
}

with open("demo.json","w") as f:
  json.dump(d,f)
```

# Pickling
* Pickling is the process whereby a Python hierarchy is converted into a byte stream , and unpickling is the inverse operation,whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy.

```python
class Person1:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def display_info(self):
    print("Hi my name is",self.name,'and I am',self.age,"year old")
```

```python
p = Person1("nitish",32)
```

```python
# Pickle dump
import pickle
with open("person.pk1","wb") as f:
  pickle.dump(p,f)
```
