```python
class Student:
  school = "New english school dorlewadi"

  def __init__(self,name):
    self.name = name

s1 = Student("Aniruddha")
s2 = Student("Priya")

print(s1.name)  # instance variables
print(s2.name)  # instance variables

print(s1.school)  # class variables
print(s2.school)  # class variables
```

```python
class Bank:
  def __init__(self,name,balance):
    self.name = name
    self.__balance = balance

  def deposit(self,amount):
    self.__balance = self.__balance + amount

  def withdraw(self,amount):
    self.__balance = self.__balance - amount

b1 = Bank("aniruddha",1000)
print(b1.name)
```

```python
class Student:
  def __init__(self,name):
    self.__name = name

  # Getter
  def get_name(self):
    return self.__name

  # setter
  def set_Name(self,new_name:str):
    self.__name = new_name


s1 = Student("Poonam")
print(s1.get_name())
```

```python
class Person:
  def __init__(self,age):
    self._age = age

  def get_age(self):
    return self._age

  def set_age(self,value):
    if value < 0:
      print("Age must be non-negative")
    self._age = value

p = Person(22)
print(p.get_age())
p.set_age(25)

print(p.get_age())
p.set_age(-5)
```

```python
class Student:
  def __init__(self,name):
    self.__name = name

  @property
  def name(self):
    print("Calling")
    return self.__name

s1 = Student("Poonam")
print(s1.name)
```
